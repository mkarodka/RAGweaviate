```python
def flatten_course_data(course_data):
    return {
        "courseName": course_data['Course']['name'],
        "courseDescription": course_data['Course']['description'],
        "courseFacultyName": course_data['Course']['faculty']['name'],
        "courseFacultyOfficeHours": course_data['Course']['faculty']['officeHours'],
        "courseOfficeHours": course_data['Course']['officeHours'],
        "requiredTexts": ', '.join(course_data['Course']['requiredTexts']),
        "learningOutcomes": ', '.join(course_data['Course']['learningOutcomes']),
        "attendancePolicy": course_data['Course']['attendancePolicy'],
        "classSchedule": course_data['Course']['classSchedule'],
        "dueDates": course_data['Course']['dueDates'],
        "prerequisites": ', '.join(course_data['Course']['prerequisites'])
    }

courses_data_flat = [
    flatten_course_data(cis_667),
    flatten_course_data(ist_718),
    flatten_course_data(ben_687)
]
```

```python
flat_schema = StructType([
    StructField("courseName", StringType(), True),
    StructField("courseDescription", StringType(), True),
    StructField("courseFacultyName", StringType(), True),
    StructField("courseFacultyOfficeHours", StringType(), True),
    StructField("courseOfficeHours", StringType(), True),
    StructField("requiredTexts", StringType(), True),  
    StructField("learningOutcomes", StringType(), True),  
    StructField("attendancePolicy", StringType(), True),
    StructField("classSchedule", StringType(), True),
    StructField("dueDates", StringType(), True),
    StructField("prerequisites", StringType(), True)  
])
df_flat = spark.createDataFrame(courses_data_flat, schema=flat_schema)
df_flat.printSchema()
```
